```python
path='https://s3.us.cloud-object-storage.appdomain.cloud/cf-courses-data/CognitiveClass/DP0701EN/version-2/Data-Collisions.csv'
df = pd.read_csv(path, dtype='unicode')
df.head()
```

```python
#changing column names where needed
df.rename(columns={'INATTENTIONIND':'Distracted'}, inplace=True)
df.head()
```

```python
#finding missing data
df.isnull()
```

```python
#drop rows with missing values from the latitude and longitude columns
df.dropna(axis=0, how='any', thresh=None, subset=['X', 'Y'], inplace=True)
```

```python
#Dropping rows with missing data from other columns where missing data can't be replaced
df.dropna(axis=0, how='any', thresh=None, subset=['JUNCTIONTYPE','WEATHER','ROADCOND','LIGHTCOND'], inplace=True)
```

```python
#Selecting collisions where light conditions were not a factor
clearLight = ['Dark - Street Lights On', 'Daylight','Dusk']
df.LIGHTCOND.isin(clearLight)


df = df[df.LIGHTCOND.isin(clearLight)]


df.LIGHTCOND.unique()
```

```python
#Selecting collisions where road conditions were not a factor
clearRoad = ['Dry']
df.ROADCOND.isin(clearRoad)


df = df[df.ROADCOND.isin(clearRoad)]


df.ROADCOND.unique()
```

```python
#mapping severe collisions where driver speeding and distracted were only causes
#install Folium
!conda install -c conda-forge folium=0.5.0 --yes
import folium
print('Folium installed and imported!')
```

```python
# create map and display it
seattle_map = folium.Map(location=[47.608013,  -122.335167], zoom_start=11)

# display the map of San Francisco
seattle_map
```

```python
# instantiate a feature group for the collisions in the dataframe
collisions = folium.map.FeatureGroup()

# loop through the collisions and add each to the incidents feature group
for lat, lng, in zip(df.Y, df.X):
    collisions.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.6
        )
    )

# add collisions to map
seattle_map.add_child(collisions)
```

```python
#plotting collisions in clusters
from folium import plugins
# let's start again with a clean map
seattle_map = folium.Map(location=[47.608013,  -122.335167], zoom_start=11)

# instantiate a mark cluster object for the incidents in the dataframe
collisions = plugins.MarkerCluster().add_to(seattle_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, in zip(df.Y, df.X):
    folium.Marker(
        location=[lat, lng],
        icon=None,
    ).add_to(collisions)

# display map
seattle_map
```
